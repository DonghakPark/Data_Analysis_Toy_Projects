```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import xgboost as xgb
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import plot_importance
%matplotlib inline
```

```python
train = pd.read_csv('./Data/train.csv', index_col=0)
```

```python
colnames = ['u', 'g', 'r', 'i', 'z', 'redshift', 'dered_u', 'dered_g', 'dered_r',
       'dered_i', 'dered_z', 'nObserve', 'nDetect', 'airmass_u', 'airmass_g',
       'airmass_r', 'airmass_i', 'airmass_z']
X, y = train[colnames], train['class']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=777)
```

```python
xgb_model = xgb.XGBClassifier(seed = 777,
                           learning_rate=0.055,
                           colsample_bytree=0.7232334448672797,
                           gamma=0.0661895281603577,
                           reg_alpha=3.54036580825445,
                           reg_lambda=0.10293226563406928,
                           subsample=0.6849549260809972,
                           max_depth=9,
                           tree_method='gpu_hist',
                           n_estimators=500)
xgb_pred = xgb_model.fit(X_train,y_train).predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred)

print("Acc : {}".format(xgb_accuracy))
```

```python
xgb_model = xgb.XGBClassifier(tree_method='gpu_hist')
xgb_pred = xgb_model.fit(X_train,y_train).predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred)

print("Acc : {}".format(xgb_accuracy))
```

```python
import xgboost

xgb_reg = xgboost.XGBRegressor()
xgb_pred = xgb_reg.fit(X_train,y_train).predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred)

print("Acc : {}".format(xgb_accuracy))
```

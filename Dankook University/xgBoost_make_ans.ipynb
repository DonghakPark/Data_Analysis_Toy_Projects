```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import xgboost as xgb
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import plot_importance
%matplotlib inline
```

```python
train = pd.read_csv('./Data/train.csv', index_col=0)
test = pd.read_csv('./Data/test.csv', index_col= 0)
sample_submission = pd.read_csv('./Data/sample_submission.csv', index_col=0)
```

```python
train_x = train.drop(columns='class', axis=1)
train_y = train['class']
test_x = test
```

```python
forest = xgb.XGBClassifier(seed = 777,
                           learning_rate=0.055,
                           colsample_bytree=0.7232334448672797,
                           gamma=0.0661895281603577,
                           reg_alpha=3.54036580825445,
                           reg_lambda=0.10293226563406928,
                           subsample=0.6849549260809972,
                           max_depth=9,
                           tree_method='gpu_hist',
                           n_estimators=500)
forest.fit(train_x, train_y)
y_pred = np.argmax(forest.predict_proba(test_x), axis =1)
```

```python
submission = pd.DataFrame(data=y_pred, columns=sample_submission.columns, index=sample_submission.index)
submission.to_csv("xgb_submission_0825_3.csv", index=True)

```
